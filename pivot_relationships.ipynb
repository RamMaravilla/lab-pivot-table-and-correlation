```python
import pandas
import numpy
```

```python
vehicles = pandas.read_csv('./vehicles.csv')
```

```python
vehicles.pivot_table(index=['Vehicle Class'])
```

```python
vehicles.pivot_table(index=["Vehicle Class"], columns=["Cylinders"], fill_value=0)
```

```python
vehicles.pivot_table(index=["Vehicle Class"], values=["Combined MPG"], aggfunc=numpy.mean)
```

```python
vehicles.pivot_table(index=["Vehicle Class"], values=["Combined MPG", "CO2 Emission Grams/Mile"])
```

```python
vehicles.pivot_table(index=["Vehicle Class"], columns=["Cylinders"], values=["Combined MPG"], fill_value=0,aggfunc=numpy.std)

```

```python
vehicles['City MPG'].corr(vehicles['Highway MPG'])

```

```python
vehicles.corr()
```

```python
vehicles.corr(method='spearman')
```

```python
vehicles['City MPG'].corr(vehicles['Highway MPG'], method="kendall")
```

```python
vehicles.corr(method='kendall')
```
